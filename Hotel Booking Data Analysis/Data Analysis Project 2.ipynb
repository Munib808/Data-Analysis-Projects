# Data Analysis And Visulization

```python
df["is_canceled"].value_counts(normalize=True)
```

```python
ax = sns.countplot(x="is_canceled",data=df,edgecolor="black")
for i in ax.containers:
    ax.bar_label(i)
```

```python
ax = sns.countplot(x="is_canceled",hue="hotel",data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
```

```python
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
Resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
City_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
City_hotel
```

```python
plt.figure(figsize=(25,8))
plt.plot(Resort_hotel.index,Resort_hotel["adr"],label="Resort_hotel")
plt.plot(City_hotel.index,City_hotel["adr"],label="City_hotel")
plt.legend()
plt.show()
```

```python
df["month"] = df["reservation_status_date"].dt.month
```

```python
try:
   plt.figure(figsize=(10,7))
   ax1 = sns.countplot(x="month",hue="is_canceled",data=df)
   legend_labels,_ = ax1.get_legend_handles_labels()
   ax1.legend(bbox_to_anchor=(1,1))
   plt.legend(["not canceled","canceled"])
   plt.show()
except:
    print("NOne")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="month",y="adr",data=df[df["is_canceled"] == 1].groupby(["month"])["adr"].sum().reset_index())
```

```python
canceled_data = df[df["is_canceled"] == 1]
top_10 = canceled_data["country"].value_counts()[:10]
plt.figure(figsize=(10,7))
# Create the pie chart
plt.pie(top_10.values, autopct="%1.1f%%", labels=top_10.index)

# Display the chart
plt.show()
```

```python
cancel = df[df["is_canceled"] == 1].groupby("reservation_status_date")[["adr"]].mean()
cancel.reset_index(inplace=True)
not_cancel = df[df["is_canceled"] == 0].groupby("reservation_status_date")[["adr"]].mean()
not_cancel.reset_index(inplace=True)
cancel.sort_values(by="reservation_status_date",ascending = True,inplace=True)
not_cancel.sort_values(by="reservation_status_date",ascending = True,inplace=True)
plt.figure(figsize=(20,6))
plt.plot(not_cancel["reservation_status_date"],not_cancel["adr"],label="not_cancel")
plt.plot(cancel["reservation_status_date"],cancel["adr"],label="cancel")

plt.legend()
plt.show()
```

```python
c = cancel[(cancel["reservation_status_date"] > "2016") & (cancel["reservation_status_date"] < "2017-09")]
n = not_cancel[(not_cancel["reservation_status_date"] > "2016") & (not_cancel["reservation_status_date"] < "2017-09")]
```

```python
plt.figure(figsize=(20,6))
plt.plot(n["reservation_status_date"],n["adr"],label="not_cancel")
plt.plot(c["reservation_status_date"],c["adr"],label="cancel")

plt.legend()
plt.show()
```
